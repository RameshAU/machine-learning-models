```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset=pd.read_csv('data.csv')
dataset.head()
```

```python
dataset['B']=dataset['B'].astype(np.int64)
dataset['S']=dataset['S'].astype(np.int64)
dataset['ST']=dataset['ST'].astype(np.int64)
dataset['MC']=dataset['MC'].astype(np.int64)
dataset['PF']=dataset['PF'].astype(np.int64)
dataset['HD']=dataset['HD'].astype(np.int64)
dataset['BS']=dataset['BS'].astype(np.int64)
dataset['D']=dataset['D'].astype(np.int64)
dataset.info()
```

```python
X = dataset.drop(['PPV','Blast NO','B','S'],axis=1)
y = dataset['PPV']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y,train_size=0.8, test_size = 0.2, random_state = 42)
```

```python
LR = LinearRegression()
LR.fit(x_train,y_train)
```

```python
y_prediction =  LR.predict(x_test)
```

```python
y_prediction
```

```python
r2_score=r2_score(y_test,y_prediction)
mse=mean_squared_error(y_test,y_prediction)
mae=mean_absolute_error(y_test,y_prediction)
```
